### 固定真实坐标系中观测者的位置，共形坐标系中观测者位置随$t_c$的变化关系

```python
# 真实坐标系 - 观测者位置
r_obs = 1e5
# 时间项
tc = np.linspace(0,15,1000)
```

```python
# 吸积率 0 ~ 0.0625
mu = 1e-3
# 更新mu
vbh.update_mu(mu)

# 更新文件路径 - mu或theta0发生变化时必须运行本函数
vbh.update_filepath()

# 初始化 共形坐标系 - 观测者位置
rcobs_1 = np.zeros_like(tc)
for idx in np.arange(len(tc)):
    # 变换至共形坐标系的观测位置
    rcobs_1[idx] = vbh.get_r2R(r_obs, tc[idx]*vbh.r0)
rcobs_1[rcobs_1<vbh.Rm] = vbh.Rm
Rp1 = vbh.Rp
```
